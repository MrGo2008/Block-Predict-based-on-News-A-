```python
'''
股市数据经过检查，没有需要进一步清洗后进行处理
股市数据处理的总体思路分为两步：
第一步：按照3,5,7,9的划窗宽度，获取简单的移动平均值，最大值，方差，标准差，最小值
第二步：将新特征与原数据合并后，再处理，思路就是某日以前的数据集合，预测未来三天的股市涨的概率

'''
#对所选数列（'open', 'low', 'high', 'close','change', 'pct_change', 'vol', 'amount', 'pe', 'pb'）进行初步的特征提取。
from multiprocessing import Pool
from tqdm import tqdm

return_features = ['open', 'low', 'high', 'close',
       'change', 'pct_change', 'vol', 'amount', 'pe', 'pb']#20190617用于与新数据合并的特征
def create_lag(df_code,windows=[3,5,7,9],shift_size=1):
    code = df_code['ts_code'].unique()
    
    for col in tqdm(return_features):#将所有的特征，按照划窗方式，获取划窗均值等特征
        for window in windows:
            rolled = df_code[col].shift(shift_size).rolling(window=window)#移动划窗取值
            lag_mean = rolled.mean()#均值
            lag_max = rolled.max()#最大值
            lag_min = rolled.min()#最小值
            lag_std = rolled.std()#标准差
            lag_var = rolled.var()#20190601#方差
            df_code['%s_lag_%s_mean'%(col,window)] = lag_mean
            df_code['%s_lag_%s_max'%(col,window)] = lag_max
            df_code['%s_lag_%s_min'%(col,window)] = lag_min
            df_code['%s_lag_%s_std'%(col,window)] = lag_std#20190601
            df_code['%s_lag_%s_var'%(col,window)] = lag_var#20190601
    return df_code.fillna(-1)

def generate_lag_features(df,windows = [3,5,7,9]):
    features = ['open', 'low', 'high', 'close',
       'change', 'pct_change', 'vol', 'amount', 'pe', 'pb']
    
    assetCodes = df['ts_code'].unique()
    print(assetCodes)
    all_df = []
    df_codes = df.groupby('ts_code')
    df_codes = [df_code[1][['trade_date','ts_code']+return_features] for df_code in df_codes]
    print('total %s df'%len(df_codes))
    pool = Pool(4)#选取4个进程
    all_df = pool.map(create_lag, df_codes)#全部扔进进程开始计算
    new_df = pd.concat(all_df)  #将结果合并
    new_df.drop(return_features,axis=1,inplace=True)
    pool.close()
    
    return new_df
```

```python
##数据按日期求历史数据的均值最大值，最小值，均值，方差等，注意这里也包含了对前面产生的划窗特征再进行特征提取
#增加了数据的傅里叶变换特征。
def get_new2(demo_df,sta_idx):
    base_cols = ['ts_code', 'trade_date', 'name']
    cols = [col for col in market_train_df.columns if col not in base_cols]
    from tqdm import tqdm
    mini_total_df = pd.DataFrame()
    for i in tqdm(range(sta_idx,demo_df.shape[0])):
    # for i in tqdm(range(0,demo_df.shape[0])):

        df = demo_df.loc[:i, cols] 
'''
历史数据预测未来三天，所以最后两天的数据的label直接填充-1,因为是在初赛基础上改过来的，当时还是预测五天，所以
多的直接添加-1,最后对多的两列也不会使用
'''       
        mini_df = pd.DataFrame()
        try:
            mini_df.loc[0,'target0'] =demo_df.loc[i, 'y']
            mini_df.loc[0,'target1'] =demo_df.loc[i+1, 'y']
            mini_df.loc[0,'target2'] =demo_df.loc[i+2, 'y']
        except:
            mini_df.loc[0,'target0'] = -1
            mini_df.loc[0,'target1'] = -1
            mini_df.loc[0,'target2'] = -1

        for col in cols:增加各历史数据傅里叶变换特征
            xc = pd.Series(df[col].values)
            zc = np.fft.fft(xc)
            realFFT=np.real(zc)
            imagFFT = np.imag(zc)
            mini_df[col+'Rmean'] = realFFT.mean()
            mini_df[col+ 'Rstd'] = realFFT.std()
            mini_df[col+ 'Rmax'] = realFFT.max()
            mini_df[col+ 'Rmin'] = realFFT.min()
            mini_df[col+'Rmean_20'] = realFFT[-20:].mean()#第五轮新增
            mini_df[col+ 'Rstd_20'] = realFFT[-20:].std()#第五轮新增
            mini_df[col+ 'Rmax_20'] = realFFT[-20:].max()#第五轮新增
            mini_df[col+ 'Rmin_20'] = realFFT[-20:].min()#第五轮新增
            mini_df[col+'Imean'] = imagFFT.mean()
            mini_df[col+ 'Istd'] = imagFFT.std()
            mini_df[col+ 'Imax'] = imagFFT.max()
            mini_df[col+ 'Imin'] = imagFFT.min()
            mini_df[col+'Imean_20'] = imagFFT[-20:].mean()#第五轮新增
            mini_df[col+ 'Istd_20'] = imagFFT[-20:].std()#第五轮新增
            mini_df[col+ 'Imax_20'] = imagFFT[-20:].max()#第五轮新增
            mini_df[col+ 'Imin_20'] = imagFFT[-20:].min()
            mini_df[col+'mini_mean'] = df[col].mean()
            mini_df[col+'mini_max'] = df[col].max()
            mini_df[col+'mini_min'] = df[col].min()
            mini_df[col+'mini_std'] = df[col].std()
            mini_df[col+'mini_var'] = df[col].var()#20190601
            mini_df[col+'mini_max_min'] = df[col].max()-df[col].min()

        mini_total_df = pd.concat([mini_total_df,mini_df])
    return mini_total_df
```

```python
#搭建双向lstm网络
import torch.nn as nn
from torch.nn import functional as F
class NeuralNet(nn.Module):
    def __init__(self, input_dim,hidden,output_dim):
        super(NeuralNet, self).__init__()
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.lstm1 = nn.LSTM(input_dim, hidden, bidirectional=True, batch_first=True)
        self.linear1 = nn.Linear(hidden*2, int(hidden/2))
        self.linear2 = nn.Linear(int(hidden/2), int(hidden/4))
        self.dropout = nn.Dropout(0.3)#评测次数有限，不敢尝试更多的值
        self.relu = nn.ReLU()
        self.output = nn.Linear(int(hidden/4),output_dim)
        
    def forward(self, x):
        out, _ = self.lstm1(x)
        out = torch.mean(out, 1)
        out = self.linear1(out)
        out = self.linear2(out)
        out = self.dropout(out)
        out = self.relu(out)
        h_conc_linear  = self.output(out)
        
        return h_conc_linear
```

```python
# 初始化网络参数
import torch
model_stock = NeuralNet(4642, 1024, 3).cuda()
optimizer = torch.optim.Adam(model_stock.parameters(), lr=0.001)
loss_F = torch.nn.MSELoss()
```

```python
#100个epoch，34个板块文件遍历一次为一个epoch，比较简陋，但是核心东西都是有了的。
for epoch in range(100):
    print("Epoch:%d"%epoch)# 数据集只迭代一次
    for code in market_train_df['ts_code'].unique():
        mini_total_df = pd.read_csv('/home/kesci/work/extra/%s.csv'%code)
        for col in mini_total_df.columns:
            mini_total_df[col] = (mini_total_df[col]-mini_total_df[col].min())/(mini_total_df[col].max()-mini_total_df[col].min()+0.1)
        X_train,Y_train,X_val,Y_val,X_test,Y_test = get_data(mini_total_df)
        # print(X_train.shape)
        out = model_stock(X_train).cuda()
        loss=loss_F(out,Y_train).cuda() # 计算loss
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        if epoch%10==9: # 每10步，计算精度
            print('Saving Model')
            if epoch<100:
                save_model_stock(model_stock)
            if epoch>100:#可以尝试更多迭代次数，为防止保存过拟合模型，所以如果需要更多尝试，可以保存两个模型
                save_model_stock(model_stock,PATH='/home/kesci/work/models/stock_model_2.model')
            with torch.no_grad():
                test_pred=model_stock(X_val).cuda()
                loss=loss_F(test_pred,Y_val).cuda()
                print(loss)
```

```python
'''
接下来是一些列函数，主要的方法就是，用Jieba对每日的新闻content进行分词，并用jieba——analysis进行关键词提取，
每个关键词通过两个词向量模型获得300维+300维共600维的词向量（我用了人民日报词向量和维基百科词向量模型，由于
只允许搭载三个文件，不然可以试试更多的词向量模型），每日新闻我们提取前599个关键词形成599x600的词向量矩阵，再把
所有关键词的词向量进行融合为一个600维的词向量，与前者融合就形成了600×600维的词向量矩阵，前七天的新闻数据按权重
形成7×600×600的一个训练数据，未来三天的股市的涨跌情况（也就是y）形成102维的数据作为label进行训练。
'''
from tqdm import tqdm
from collections import Counter
import jieba
import re
# content = news_train_df[news_train_df['date']==20140414].content.values
def content_to_words(content,double):
    r1 = u'[a-zA-Z0-9’!"#$%&\'()*+,-./:;<=>?@，。?★、…【】《》？“”‘’！[\\]^_`{|}~]+'
    
    all_word=[]#所有关键词
    for sentence in content:
        sentence_word=[]
        try:
            sentence=re.sub(r1, '',sentence)#清洗句子
            # words = jieba.cut_for_search(sentence)
            words =jieba.analyse.extract_tags(sentence, topK=-1, withWeight=False)#每个句子提取关键词
            all_word.extend(word for word in words if word not in stopwords)
        except:
            continue
    counter = Counter(all_word)#词频
    if double:
        count_pair1 = counter.most_common(len(all_word))#所有词语
        count_pair2 = counter.most_common(599)#前599个词语
    else:
        count_pair = counter.most_common(300)#如果只有一个模型可考虑这个
    return all_word,count_pair1,count_pair2
    
def get_vector(news,double):
    #词向量再次集合，组成训练数据
    words,count1,count2 = content_to_words(news,double)
    
    if double:
        cab = np.zeros([600,600])
        idx = 0
        for word,fre in count2:
            try:
                cab[idx,:300]=model1[w]
                cab[idx,300:]=model2[w]
                idx+=1
            except:
                cab[idx,:]=np.random.randn(600)#如果不能找到对应词语的词向量，就随机填充
                idx+=1
                continue
        for w,f in count1:
            try:
                cab[599,:300]+=model1[w]
                cab[599,300:]+=model2[w]
            except:
                cab[599,:]+=np.random.randn(600)
                idx+=1
                continue
        cab[599,:]=cab[599,:]/(len(count1)+0.1)#为防止出现除0
        return cab
```

```python
import torch
import time
class BasicModule(torch.nn.Module):
    # 封装了nn.Module， 主要是提供了save和load两个方法

    def __init__(self):
        super(BasicModule, self).__init__()
        self.model_name = str(type(self))  # 模型的默认名字

    def get_optimizer(self, lr1, lr2=0, weight_decay=0):
        embed_params = list(map(id, self.embedding.parameters()))
        base_params = filter(lambda p: id(p) not in embed_params, self.parameters())
        optimizer = torch.optim.Adam([
            {'params': self.embedding.parameters(), 'lr': lr2},
            {'params': base_params, 'lr': lr1, 'weight_decay': weight_decay}
        ])
        return optimizer
```

```python
'''
网络结构采取textcnn的基本结构，主要是在最后一层池化的范围有所增加，理论上扩大了每个节点的感知域，这一点
让预测结果效果更加明显。
'''
# text cnn
import torch
import torch.nn as nn
import numpy as np
kernal_sizes = [1, 2, 3, 4, 5]
class TextCNN(BasicModule):
    def __init__(self, vectors=None):
        super(TextCNN, self).__init__()
        embedding_dim = 7,
        kernel_num = 1024,
        kernel_size = 7
        linear_hidden_size = 512
        convs = [
            nn.Sequential(
                nn.Conv1d(in_channels=600,
                          out_channels=1024,
                          kernel_size=7),
                nn.BatchNorm1d(1024),
                nn.ReLU(inplace=True),

                nn.Conv1d(in_channels=1024,
                          out_channels=1024,
                          kernel_size=7),
                nn.BatchNorm1d(1024),
                nn.ReLU(inplace=True),
                nn.MaxPool1d(kernel_size=(600 - 7*k + 2))
            )
            for k in kernal_sizes  # 池化层划窗
        ]
        self.convs = nn.ModuleList(convs)

        self.fc = nn.Sequential(
            nn.Linear(5120, 512),
            nn.BatchNorm1d(512),
            nn.ReLU(inplace=True),
            nn.Linear(512, 102)
        )
    def forward(self, inputs):
        # embeds = self.embedding(inputs)  # seq * batch * embed
        # 进入卷积层前需要将Tensor第二个维度变成emb_dim，作为卷积的通道数
        # conv_out = [conv(embeds.permute(1, 2, 0)) for conv in self.convs]
        conv_out = [conv(inputs) for conv in self.convs]
        conv_out = torch.cat(conv_out, dim=1)
        flatten = conv_out.view(conv_out.size(0), -1)
        logits = self.fc(flatten)
        logits = torch.sum(logits, 0, True)  # 多个预测值合并
        return logits
```
